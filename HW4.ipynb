```python
"""
Created April 26, 2020
Steven Weber
3d plotting of interesting functions
"""
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D # for 3d surface plot



    


# standard Gaussian (normal) PDF
def f_gaussian(x,y):
	return (1/(2*np.pi))*np.exp((-x**2 + y**2)/2)

# taken from: https://www.benjoffe.com/code/tools/functions3d/examples
def f_ripple(x,y):
	return np.sin(x**2+y**2)

# taken from: https://www.benjoffe.com/code/tools/functions3d/examples
def f_bumps(x,y):
	return np.sin(x)*np.cos(y)

# taken from: https://www.benjoffe.com/code/tools/functions3d/examples
def f_pyramid(x,y):
	return 1- abs(x+y)-abs(x-y)

# taken from: https://www.benjoffe.com/code/tools/functions3d/examples
def f_cross(x,y):
	return .1 - np.sign(np.sign((x*12)**2 - 9)-1 + np.sign((y*12)**2 -9)-1)/2

# taken from: https://www.benjoffe.com/code/tools/functions3d/examples
def f_cross2(x,y):
	return np.exp(-(2*x)**2 * (2*y)**2)

# taken from: https://www.benjoffe.com/code/tools/functions3d/examples
def f_windmill(x,y):
    return np.sign(x*y)*np.sign(1-(9*x**2)+(9*y**2))

"""
plot3d: create a 3-dimensional surface plot
input mx: x-axis values
input my: y-axis values
input mz: z-axis values
input xlbl: x-axis label
input ylbl: y-axis label
input zlbl: z-axis label
input title: plot title
input filename: save figure as
"""
def plot3d(mx, my, mz, xlbl, ylbl, zlbl, title, filename):
    plt.figure()
    ax = plt.axes(projection="3d")
    ax.plot_surface(mx , my , mz , cmap='viridis')
    ax.set_xlabel (xlbl)
    ax.set_ylabel (ylbl)
    ax.set_zlabel (zlbl)
    ax.set_title (title)
    plt.savefig(filename)
    return plt.show()

           

              

"""
set range of values for x,y axes
set vnum points uniformly between vmin and vmax
use meshgrid to create x-axis and y-axis points mx, my
"""
vmin,vmax,vnum = -2.0, +2.0, 101
v = np.linspace(vmin,vmax,vnum)
mx,my = np.meshgrid(v,v)

# Bivariate normal (Gaussian) PDF
mz_gaussian = f_gaussian(mx,my)
#print(mz_gaussian)
plot3d(mx, my, mz_gaussian, 'x', 'y', 'f(x,y)', 'gaussian', 'HW4Solution-1.pdf')

# "ripple" function
mz_ripple = f_ripple(mx,my)
plot3d(mx, my, mz_ripple, 'x', 'y', 'f(x,y)', 'ripple', 'HW4Solution-2.pdf')

# "bumps" function
mz_bumps = f_bumps(mx,my)
plot3d(mx, my, mz_bumps, 'x', 'y', 'f(x,y)', 'bumps', 'HW4Solution-3.pdf')

# "pyramid" function
mz_pyramid = f_pyramid(mx,my)
plot3d(mx, my, mz_pyramid, 'x', 'y', 'f(x,y)', 'pyramid', 'HW4Solution-4.pdf')

# "cross" function
mz_cross = f_cross(mx,my)
plot3d(mx, my, mz_cross, 'x', 'y', 'f(x,y)', 'cross', 'HW4Solution-5.pdf')

# "cross" function
mz_cross2 = f_cross2(mx,my)
plot3d(mx, my, mz_cross2, 'x', 'y', 'f(x,y)', 'cross2', 'HW4Solution-6.pdf')

# "windmill" function
mz_windmill = f_windmill(mx,my)
plot3d(mx, my, mz_windmill, 'x', 'y', 'f(x,y)', 'windmill', 'HW4Solution-7.pdf')
```
